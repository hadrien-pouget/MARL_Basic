- Naive and lola updates act nearly the same in one-shot games. I've observed this on all the one-shot games I've tried
- Both actually seem to do pretty well when policies are trained together, but this gets much worse if not, as expected (AA's paired with BB's is really bad, AA or BB paired with AB is middling bad). In this case, the nash equilibria are pretty good though.
- AB is the best, AA and BB are ok, BA is the worst.
